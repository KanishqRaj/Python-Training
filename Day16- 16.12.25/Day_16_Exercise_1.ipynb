#Task 1 - Read the CSV file using sparkContext.textFile and display the first 5 records.

```python
rdd = spark.sparkContext.textFile("call_logs.csv")
first5 = rdd.take(5)
for row in first5:
  print(row)
```

#Task 2 - Remove the header row and create a clean RDD containing only data rows.




```python

# rdd = spark.sparkContext.textFile("call_logs.csv")
header = rdd.first()
clean_rdd = rdd.filter(lambda row: row != header)
clean_rdd.take(5)
```

#Task 3 - Split each row into individual fields using a delimiter.

```python
split_rdd = clean_rdd.map(lambda row : row.split(','))
split_rdd.take(5)
```

#Task 4 - Calculate the total call cost per city.

```python

city_cost_rdd = split_rdd.map(lambda parts: (parts[3], float(parts[6])))
total_cost_per_city = city_cost_rdd.reduceByKey(lambda a, b: a + b)
total_cost_per_city.collect()
```

#Task 5 - Identify the city with the highest total call cost.

```python
max_cost_city = total_cost_per_city.reduce(lambda x , y: x if x[1] > y[1] else y )
print(max_cost_city)
```
